```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1= pd.read_csv('tienda_1 .csv')
df2= pd.read_csv('tienda_2.csv')
df3= pd.read_csv('tienda_3.csv')
df4= pd.read_csv('tienda_4.csv')
```

```python
df1.info()
```

```python
ingreso_total1 = df1['Precio'].sum()
ingreso_total2 = df2['Precio'].sum()
ingreso_total3 = df3['Precio'].sum()
ingreso_total4 = df4['Precio'].sum()
```

```python

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
plt.bar(tiendas, [ingreso_total1, ingreso_total2, ingreso_total3, ingreso_total4])

plt.xlabel('Tienda')
plt.ylabel('Precio')
plt.title('Gráfico de Barras')
plt.show()
```

```python
grupo_tienda1 = df1.groupby('Categoría del Producto').size()
grupo_tienda2 = df2.groupby('Categoría del Producto').size()
grupo_tienda3 = df3.groupby('Categoría del Producto').size()
grupo_tienda4 = df4.groupby('Categoría del Producto').size()
```

```python
envio_promedio1 = df1['Costo de envío'].mean()
envio_promedio2 = df2['Costo de envío'].mean()
envio_promedio3 = df3['Costo de envío'].mean()
envio_promedio4 = df4['Costo de envío'].mean()

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_envio = [envio_promedio1, envio_promedio2, envio_promedio3, envio_promedio4]

plt.bar(tiendas, costos_envio, color='skyblue')
plt.xlabel('Tiendas')
plt.ylabel('Costo de Envío Promedio')
plt.title('Costo de Envío Promedio por Tienda')

plt.tight_layout()
plt.show()
```
